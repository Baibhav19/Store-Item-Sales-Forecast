```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('data/train.csv', parse_dates=True, index_col='date', infer_datetime_format=True)
```

### Sum of sale of items on a day on all stores.

```python
df_gby_item = df.groupby(['date','item'])['sales'].sum().unstack()
```

```python
ax = df_gby_item[1].plot(figsize=(12,6), legend=True)
ax.set(xlabel='date', ylabel='Sales')
df_gby_item[2].plot(legend=True)
df_gby_item[3].plot(legend=True);
```

### Starting 3 items sales show a similar trend. Analyzing and forecasting for a single item

```python
item1_df = df_gby_item[[1]].copy()
```

```python
item1_df.dropna(inplace=True)
```

```python
results = seasonal_decompose(item1_df)
ax = results.seasonal[:30].plot(figsize=(12,6))
ax.autoscale(axis='x',tight=True)
```

```python
results = seasonal_decompose(item1_df)
ax = results.trend.plot(figsize=(12,6))
ax.autoscale(axis='x',tight=True)
```

### Data shows a clear weekly and yearly trend.

### Now Checking whether data is stationary or not using adfuller test

```python
adfuller(item1_df)
```

#### p-value is 0.03, so we can reject null hypothesis. Data is Stationary.

```python
item1_df.loc[:,'day_of_week'] = item1_df.index.day_name()
```

```python
item1_df.rename(columns={1:"Sales"}, inplace=True)
```

```python
df1 = pd.DataFrame(pd.get_dummies(item1_df['day_of_week'], prefix='day', dtype=int), index=item1_df.index)
```

```python
item1_df_fin = pd.concat([item1_df, df1], axis=1)
```

```python
fig = plt.gcf()
fig.set_size_inches(12,6)
sns.violinplot(data=item1_df_fin,x='day_of_week',y='Sales');
```

### For yearly periodic data we can use fourier series to estimate this period 

```python

for i in range(1,5):
    item1_df_fin_seas = item1_df_fin.copy()
    for j in range(1,i+1):
        
        item1_df_fin_seas['sin_' + str(j)] = np.sin(2 * np.pi * j * item1_df_fin_seas.index.dayofyear/365.25)
        item1_df_fin_seas['cos_' + str(j)] = np.cos(2 * np.pi * j * item1_df_fin_seas.index.dayofyear/365.25)
    au = auto_arima(item1_df_fin_seas['Sales'], x=item1_df_fin_seas[[cols for cols in item1_df_fin_seas.columns if cols != 'Sales']]
                    ,seasonal=True, m=7)
    print(au.aic(), i)
    
```

### For k=2 we see low value of AIC.  

```python
item1_df_fin_seas = item1_df_fin.copy()
i=2
for j in range(1,i+1):
        
    item1_df_fin_seas['sin_' + str(j)] = np.sin(2 * np.pi * j * item1_df_fin_seas.index.dayofyear/365.25)
    item1_df_fin_seas['cos_' + str(j)] = np.cos(2 * np.pi * j * item1_df_fin_seas.index.dayofyear/365.25)
au = auto_arima(item1_df_fin_seas['Sales'], x=item1_df_fin_seas[[cols for cols in item1_df_fin_seas.columns if cols != 'Sales']],
                trace=True,seasonal=True, m=7)
```

```python
fig = plt.gcf()
fig.set_size_inches(12,6)
sns.heatmap(item1_df_fin_seas.corr());
```

```python
item1_df_fin_seas.drop('day_of_week', axis=1, inplace=True)
```

### Splitting to train and test data. Test size = 3 months 

```python
train = item1_df_fin_seas[:1736]
test = item1_df_fin_seas[1736:]
```

```python
model = SARIMAX(train['Sales'],exog=train[[cols for cols in item1_df_fin_seas.columns if cols != 'Sales']],
                order=(0,1,1),seasonal_order=(1,0,1,7),enforce_invertibility=False)
```

```python
results = model.fit()
results.summary()
```

```python
test_start=1736
test_end=1825
```

```python
predictions = results.predict(start=test_start,end=test_end,exog=test[[cols for cols in item1_df_fin_seas.columns if cols != 'Sales']]).rename('Pred_sales')
```

```python
ax = test['Sales'].plot(figsize=(12,6),legend=True)
ax.set(xlabel='date', ylabel='Sales')
predictions.plot(legend=True);
```

```python
print("Root mean square error: ",np.sqrt(mean_squared_error(test['Sales'], predictions)))
```

### Generating forecast exogenous variables data for 3 months

```python
forecast_index = pd.date_range(start=str(item1_df_fin_seas.index[-1:].values[0]),periods=90,freq='D')
item1_forecast = pd.DataFrame(index=pd.to_datetime(forecast_index))
item1_forecast['day_of_week'] = item1_forecast.index.day_name()
item1_forecast_fin = pd.DataFrame(pd.get_dummies(item1_forecast['day_of_week'], prefix='day', dtype=int), index=item1_forecast.index)

i=2
for j in range(1,i+1):
        
    item1_forecast_fin['sin_' + str(j)] = np.sin(2 * np.pi * j * item1_forecast_fin.index.dayofyear/365)
    item1_forecast_fin['cos_' + str(j)] = np.cos(2 * np.pi * j * item1_forecast_fin.index.dayofyear/365)
```

```python
model = SARIMAX(item1_df_fin['Sales'],exog=item1_df_fin_seas[[cols for cols in item1_df_fin_seas.columns if cols != 'Sales']],order=(0,1,1),seasonal_order=(1,0,1,7),enforce_invertibility=False)
results_full=model.fit()
```

```python
forecast = results_full.predict(start=1826,end=1915,exog=item1_forecast_fin).rename('Forecast Sales')
```

```python
forecast.plot(figsize=(10,5),legend=True);
```
